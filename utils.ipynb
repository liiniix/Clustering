```python
import pandas as pd
```

```python
dat = pd.read_csv("crx.data", header=None).sample(frac=1).reset_index(drop=True)
```

```python
dat.dtypes
```

```python
one_hot = pd.get_dummies(dat[4])
```

```python
del dat[4]
```

```python
dat = dat.join(one_hot)
```

```python
dat.columns = [i for i in range(len(dat.columns))]
```

```python
dat = dat.apply(pd.to_numeric, errors='coerce')
```

```python
dat.dtypes
```

```python
del dat[0]
```

```python
dat
```

```python
dat.to_csv("aaa", header=False, index=False)
```
